# Linear Regression Model

```python
import scipy.stats as stats
```

```python
import sklearn
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = data.drop(['SalePrice'] , axis = 1) ## As every column is changed to numerical data.
y = data.loc[:,'SalePrice']
```

```python
regr = linear_model.LinearRegression() 
regr.fit(X, y) 
```

```python
predictions = regr.predict(X) ## predicting for training data
```

```python
mse = mean_squared_error(y,predictions)
print('Root Mean Squared Error : %f'%np.sqrt(mse))
```

# Residual vs Fit Line

```python
plt.scatter(predictions, y)
plt.show()
```

```python
plt.scatter(predictions, y - predictions) ## Residual vs fitted
plt.show()
```

Clearly it doesn't appear like a line. It looks like a quadratic polynomial with as it is a U shape.

## Polynomial Regression Model ###

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2)  ## Converting into quadratic polynomial
X_poly = poly.fit_transform(X) 
poly_model = LinearRegression()
```

```python
poly_model.fit(X_poly,y)
```

```python
prediction_poly = poly_model.predict(X_poly)
```

```python
mse_poly = mean_squared_error(y,prediction_poly)
print('Root Mean Squared Error : %f'%np.sqrt(mse_poly))
```

```python
plt.scatter(prediction_poly, y)
plt.show()
```

```python
plt.scatter(prediction_poly, y)
plt.show()
```

# Cross Validation to reduce overfitting

```python
from sklearn.model_selection import cross_val_score, ShuffleSplit
q_scores = cross_val_score(poly_model, X_poly, y, cv = 5) ## Using stratified k fold algo which is default
print('Quadratic Regression Mean Accuracy Percentage: %f' %(q_scores.mean() * 100))
l_scores = cross_val_score(regr, X, y, cv = 5)
print('Linear Regression Mean AccuracyPercentage: %f' %(l_scores.mean() * 100))
```

As we can see Polynomial Regression model is overfit(Very Low RSME value) because of unusual mean percentage, so Linear Regression model is used to get the testing data predictions though it's RMSE is greater than quadratic model's.
